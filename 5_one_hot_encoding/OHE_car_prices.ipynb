```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("carprices.txt")
df
```

```python
dummies=pd.get_dummies(df[['Car Model']])
dummies
```

```python
new_df=pd.concat([df,dummies],axis='columns')
new_df
```

```python
new_df=new_df.drop(['Car Model','Car Model_Audi A5'],axis='columns')
new_df
```

```python
X=new_df.drop(['Sell Price($)'],axis='columns')
X
```

```python
y=new_df[['Sell Price($)']]
y
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(X,y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.predict([[86000,7,1,0]])
```

```python
model.score(X,y)
```

# using Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle['Car Model']=le.fit_transform(df['Car Model'])
dfle
```

```python
X=dfle[['Car Model','Mileage','Age(yrs)']].values
X
```

```python
y=dfle[['Sell Price($)']].values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('Car Model',OneHotEncoder(),[0])],remainder="passthrough")
```

```python
X=ct.fit_transform(X)
X
```

```python
X=X[:,1:] #removing one category 
X
```

```python

y
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,45000,4]])
```

```python
model.score(X,y)
```
