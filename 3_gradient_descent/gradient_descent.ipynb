```python
import pandas as pd
import numpy as np
import math
```

```python
df=pd.read_csv("test_score.txt")
df
```

```python
x=np.array(df[['math']])
x
```

```python
y=np.array(df[['cs']])
y
```

```python
def gradient_descent(x,y):
    m_cur=b_cur=0
    iterations=1000
    learning_rate=0.0002
    n=len(x)
    cost_previous=0
    
    for i in range(iterations):
        y_pred=m_cur * x + b_cur
        cost= (1/n) * sum ([val**2 for val in(y-y_pred)])
        md=-(2/n)* sum (x*(y-y_pred))
        bd=-(2/n)* sum( y- y_pred)
        m_cur= m_cur - learning_rate*md
        b_cur= b_cur - learning_rate*bd
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        print("m {} b {} cost {}".format(m_cur,b_cur,cost))
        
```

```python
gradient_descent(x,y)
```
