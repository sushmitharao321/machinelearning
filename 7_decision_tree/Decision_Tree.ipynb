```python
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_csv("salary.txt")
df
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
inputs
```

```python
target= df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_en']=le_company.fit_transform(inputs['company'])
inputs['job_en']=le_job.fit_transform(inputs['job'])
inputs['degree_en']=le_degree.fit_transform(inputs['degree'])
inputs
```

```python
new_input=inputs.drop(['company','job','degree'], axis='columns')
new_input
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(new_input,target,test_size=0.2)
```

```python
model.fit(new_input,target)
```

```python
model.score(new_input,target)
```
